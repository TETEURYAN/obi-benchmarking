```python
import pandas as pd
import numpy as np
import os
from pathlib import Path
import json
```

```python
path_database = Path("../database")
questions_path = []
        
for folder in list(path_database.iterdir()):
    print(folder)
    if folder.is_dir():
        questions_path.append(folder)
        
problems = []
for question_path in questions_path:
    problems.append(json.loads(Path(question_path / "problem.json").read_text(encoding="utf-8")))
```

```python
for problem in problems:
    print(problem["title"])
```

```python
file_path = "../output/results/results_gemini-3.1-pro-preview_python_zero.csv"

if os.path.exists(file_path):
    df = pd.read_csv(file_path)
    print("Arquivo carregado com sucesso!")
else:
    print(f"Erro: O arquivo '{file_path}' não foi encontrado. Verifique se o judge já terminou de rodar.")
```

```python
total_test_cases = 0

for count_test in df["total_test_cases"]:
    total_test_cases += count_test
    
print(f"Total de casos teste: {total_test_cases}")
```

```python
languages = ['python', 'cpp']
tec = ['zero', 'few']
name_models = ['gemini-3.1-pro-preview', 'gpt-5.4-2026-03-05']

results = []

for language in languages:
    for t in tec:
        for name_model in name_models:
            file_path = f"../output/results/results_{name_model}_{language}_{t}.csv"
            
            
            if os.path.exists(file_path):
                df = pd.read_csv(file_path)
                # print(f"Arquivo {file_path} carregado com sucesso!")
                
                total_test = 0
                total_judge = 0
                for count_test, judge in zip(df["total_test_cases"], df["judge_predict"]):
                    if judge != "NO CODE":
                        total_test += count_test
                        total_judge += 1
                
                linha = {}
                linha["judge"] = total_judge
                linha["model"] = name_model
                linha["Linguagem"] = language
                linha["Tecnica"] = t
                
                tempo = df['llm_code_creation_time'].sum()
                soma = df["AC"].sum()
                taxa = float(soma) / float(total_test)
                
                linha["Taxa de acerto"] = taxa
                linha["Total de acertos - Total de casos test"] = f"{soma}/{total_test}"
                linha["Tempo de criação das questões"] = tempo
                
                results.append(linha)
            else:
                print(f"Erro: O arquivo '{file_path}' não foi encontrado. Ignorando...")

print("Modelo | Técnica | Linguagem | Total judge | Acertos/Total | Taxa de acertos judge | Tempo de criação das questões")
for result in results:
    # CORREÇÃO: Usando 'result' em vez de 'linha'
    print(f"{result['model']} | {result['Tecnica']} | {result['Linguagem']} | {result['judge']}/13 | {result['Total de acertos - Total de casos test']} | {result['Taxa de acerto'] * 100:.2f}% | {result['Tempo de criação das questões']:.2f}s")
```
